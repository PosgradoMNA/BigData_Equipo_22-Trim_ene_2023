###Creando nuevas características:

####Tabla Hogares:

**1. Promedio de personas por hogar en Jalisco en cada municipio en el último año:**

```
%sql
SELECT cve_municipio, 
       sum(CASE WHEN indicador = 'Población en hogares' THEN valor ELSE 0 END) / 
       sum(CASE WHEN indicador = 'Hogares' THEN valor ELSE 0 END) AS promedio_poblacion_por_hogar,
       "Promedio de población por hogar" AS unidad_promedio_poblacion
FROM indicadoresHogares
WHERE ano = (SELECT MAX(ano) FROM indicadoresHogares) AND cve_entidad = "14"
GROUP BY cve_municipio
SORT BY promedio_poblacion_por_hogar;
```

####Tabla DENUE:

**1. Tiempo en años que llevan las compañías:**

```
%sql 
SELECT id AS id_denue, (YEAR(GETDATE()) - CAST(SUBSTR(fecha_alta,1,4) AS INT)) AS anos_de_la_compania 
FROM denue
LIMIT 10
```

**2. Tamaño de la compañía por número de empleados:**

```
%sql
SELECT id AS id_denue, per_ocu,
  CASE
    WHEN per_ocu = '0 a 5 personas' OR per_ocu = '6 a 10 personas' OR per_ocu = '11 a 30 personas' OR per_ocu = '31 a 50 personas' THEN 'Pequeña'
    WHEN per_ocu = '51 a 100 personas' OR per_ocu = '101 a 250 personas' THEN 'Mediana'
    WHEN per_ocu = '251 y más personas' THEN 'Grande'
    ELSE 'No Clasificada'
  END AS tamano_empresa
FROM denue
LIMIT 10
```

####Tabla Educación:

**1. Tasa de alfabetización por municipios y cantidad de empresas:**
Se cruza la tabla DENUE con la de Educación para obtener la cantidad de empresas en la zona y averiguar si existe alguna correlación con la tasa de alfabetización en personas mayores a 15 años.

```
# specify path
pathIndicadoresEducacion= "/FileStore/tables/educacion_load.csv"

# load as text
dataIndicadoresEducacion = spark.read.csv(pathIndicadoresEducacion) 

# show sample
dataIndicadoresEducacion.take(20)

###################################################

# read in file using csv format
dfIndicadoresEducacion = spark.read.load(pathIndicadoresEducacion,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true',
                    encoding='ISO-8859-1')

# show 20 rows
display(dfIndicadoresEducacion)
```

```
%sql 
SELECT IE.cve_municipio, IE.desc_municipio,
       SUM(CASE WHEN IE.indicador = 'Porcentaje de personas de 15 años y más alfabetas' THEN valor ELSE 0 END) AS promedio_alfabetas_mayores_15
FROM indicadoresEducacion IE
WHERE IE.ano = (SELECT MAX(ano) FROM indicadoresEducacion) AND IE.cve_entidad = "14"
GROUP BY IE.cve_municipio, IE.desc_municipio
SORT BY promedio_alfabetas_mayores_15 DESC
LIMIT 10
```

```
%sql
SELECT D.cve_mun, D.municipio, count(D.id) AS cantidad_empresas, 
       IE.promedio_alfabetas_mayores_15
FROM denue D
JOIN (
   SELECT IE.cve_municipio, IE.desc_municipio,
          SUM(CASE WHEN IE.indicador = 'Porcentaje de personas de 15 años y más alfabetas' THEN valor ELSE 0 END) AS promedio_alfabetas_mayores_15
   FROM indicadoresEducacion IE
   WHERE IE.ano = (SELECT MAX(ano) FROM indicadoresEducacion) AND IE.cve_entidad = "14"
   GROUP BY IE.cve_municipio, IE.desc_municipio
) IE 
ON D.cve_mun = IE.cve_municipio
GROUP BY D.cve_mun, D.municipio, IE.promedio_alfabetas_mayores_15
SORT BY cantidad_empresas DESC
LIMIT 50
```

####Tablas Natalidad, Mortalidad y Población:

Se usaron los datos de estas tablas para calcular la siguiente nueva característica, que corresponde a la tasa de crecimiento poblacional sin inmigrantes o tasa de crecimiento natural por cada mil habitantes en cada municipio para el año 2020:
</br>
- Tasa de crecimiento natural = [(Nacimientos - Defunciones) / Población inicial] x 1000

**1. Tasa de crecimiento natural:** = [(Nacimientos(2020) - Defunciones(2020)) / Población inicial(2020)] x 1000

**PD: El ejercicio se hace con la población del año 2020, ya que el censo no registra población del año anterior para sacar la media**

```
# specify path... Natalidad
pathIndicadoresNatalidad= "/FileStore/tables/natalidad_load.csv"

# load as text
dataIndicadoresNatalidad = spark.read.csv(pathIndicadoresNatalidad)
#---------------------------------------------------------------------
# specify path... Mortalidad
pathIndicadoresMortalidad= "/FileStore/tables/mortalidad_load.csv"

# load as text
dataIndicadoresMortalidad = spark.read.csv(pathIndicadoresMortalidad)
#---------------------------------------------------------------------
# specify path... Población
pathIndicadoresPoblacion= "/FileStore/tables/poblacion_load.csv"

# load as text
dataIndicadoresPoblacion = spark.read.csv(pathIndicadoresPoblacion) 

# show sample
#dataIndicadoresPoblacion.take(20)
#---------------------------------------------------------------------
# read in file using csv format... Natalidad
dfIndicadoresNatalidad = spark.read.load(pathIndicadoresNatalidad,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true',
                    encoding='ISO-8859-1')
#---------------------------------------------------------------------
# read in file using csv format... Mortalidad
dfIndicadoresMortalidad = spark.read.load(pathIndicadoresMortalidad,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true',
                    encoding='ISO-8859-1')
#---------------------------------------------------------------------
# read in file using csv format... PoblacionMortalidad
dfIndicadoresPoblacion = spark.read.load(pathIndicadoresPoblacion,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true',
                    encoding='ISO-8859-1')

# show 20 rows
display(dfIndicadoresPoblacion)
```

```
dfIndicadoresNatalidad.createOrReplaceTempView('indicadoresNatalidad')
dfIndicadoresMortalidad.createOrReplaceTempView('indicadoresMortalidad')
dfIndicadoresPoblacion.createOrReplaceTempView('indicadoresPoblacion')
```

```
%sql
SELECT INat.cve_municipio, INat.desc_municipio, INat.nacimientos, IM.muertes, IP.poblacion_total,
       ((INat.nacimientos - IM.muertes) / IP.poblacion_total)*1000 AS tasa_crecimiento_natural
FROM (
  SELECT cve_municipio, desc_municipio,
         CASE WHEN indicador = 'Nacimientos registrados' THEN valor ELSE 0 END AS nacimientos
  FROM indicadoresNatalidad
  WHERE cve_entidad = '14' AND ano = '2020' AND indicador = 'Nacimientos registrados'
) INat
JOIN (
  SELECT cve_municipio, desc_municipio,
         CASE WHEN indicador = 'Defunciones registradas' THEN valor ELSE 0 END AS muertes
  FROM indicadoresMortalidad
  WHERE cve_entidad = '14' AND ano = '2020' AND indicador = 'Defunciones registradas'
) IM ON INat.cve_municipio = IM.cve_municipio
JOIN (
  SELECT cve_municipio, desc_municipio,
         CASE WHEN indicador = 'Población total' THEN valor ELSE 0 END AS poblacion_total
  FROM indicadoresPoblacion
  WHERE cve_entidad = '14' AND ano = '2020' AND indicador = 'Población total'
) IP ON INat.cve_municipio = IP.cve_municipio
SORT BY tasa_crecimiento_natural DESC
LIMIT 10;
```
